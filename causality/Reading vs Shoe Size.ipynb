```python
import numpy as np
import numpy.random as random
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
```

```python
effect_reading_on_foot_size = 0
effect_age_on_foot_size = 2
effect_age_on_reading = 1

def noise():
    return (random.random() - 0.5) * 0.3

# Model:
# age -> cholesterol <- exercises <- age

ages = np.array(range(40, 150)) / 10
reading = [ z / 2 * (effect_age_on_reading + noise()) for z in ages ]
foot_size = [ z // 3 * (effect_age_on_foot_size + noise()) +
              x * (effect_reading_on_foot_size + noise()) for x, z in zip(reading, ages) ]
```

```python
plt.scatter(x = reading, y = foot_size)
plt.xlabel("reading")
plt.ylabel("foot_size")

plt.show()
```

```python
for age_group in range(10, 50, 10):
#     exercises_for_age = exercises[age_group-10:age_group]
#     cholesterol_for_age = cholesterol[age_group-10:age_group]
    circle = Ellipse((age_group / 8 + 1.5, age_group / 5), 2, 3 + age_group * 0.04, angle=80, color='r', fill=False, clip_on=False)
    plt.gca().add_artist(circle)
    
plt.scatter(x = reading, y = foot_size)
plt.xlabel("reading")
plt.ylabel("foot_size")

plt.show()
```
