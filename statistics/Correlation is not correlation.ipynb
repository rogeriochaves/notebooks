```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = list(range(-100, 100))
y = [ x ** 2 for x in x ]
plt.plot(x, y)
"correlation: %.1f" % np.corrcoef(x, y)[1][0]
```

Theoretically mutual information should be a better metric, let's see

```python
from scipy.stats import chi2_contingency

# From: https://stackoverflow.com/a/20505476
def show_MI(x, y, bins):
    c_xy = np.histogram2d(x, y, bins)[0]
    g, p, dof, expected = chi2_contingency(c_xy, lambda_="log-likelihood")
    mi = 0.5 * g / c_xy.sum()
    plt.plot(x, y)
    print("correlation: %.1f" % np.corrcoef(x, y)[1][0])
    print("mutual info:", mi)

x = list(range(-100, 100))
y = np.random.rand(200)

show_MI(x, y, bins=10)
```

```python
x = list(range(-100, 100))
y = [ x ** 2 if np.random.rand() > 0.5 else np.random.rand() * 100 ** 2 for x in x ]

show_MI(x, y, bins=10)
```

```python
x = list(range(-100, 100))
y = [ x ** 2 for x in x ]

show_MI(x, y, bins=10)
```

```python
x = list(range(-100, 100))
y = x

show_MI(x, y, bins=10)
```

```python
x = list(range(-100, 100))
y = [ x ** 3 for x in x ]

show_MI(x, y, bins=10)
```

```python
x = list(range(-100, 100))
y = [ -x for x in x ]

show_MI(x, y, bins=10)
```

```python
# IQ vs success example

x = [0] * 1000 + list(range(0, 100)) # [0] * 1000 is "dead people"
y = [ x if x < 10 else 10 for x in x ]

show_MI(x, y, bins=10)
```
