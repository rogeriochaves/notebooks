```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import math
```

```python
ys = []
for i in range(0, 100):
    y = 1 + 2 * np.random.randn()
    ys.append(y)
ys.sort()
ys

ys2 = []
for i in range(0, 100):
    y2 = 3 + 1 + math.sqrt(3) * np.random.randn()
    ys2.append(y2)
ys2.sort()
ys2

hmean = np.mean(ys)
hstd = np.std(ys)
pdf = stats.norm.pdf(ys, hmean, hstd)
plt.plot(ys, pdf)

hmean = np.mean(ys2)
hstd = np.std(ys2)
pdf = stats.norm.pdf(ys2, hmean, hstd)
plt.plot(ys2, pdf);
```
