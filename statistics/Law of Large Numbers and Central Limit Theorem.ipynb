```python
import numpy as np
import matplotlib.pyplot as plt
```

# Normal

```python
dist = np.random.normal(size=10000)
print("std", np.std(dist))
plt.hist(dist, bins=100);
```

```python
def plot_std_reduction(distribution):
    m = 500
    stds = []
    for n in range(1, m):
        dist = np.mean(distribution(n), axis=1)
        stds.append(np.std(dist))

    print("std", np.std(dist))
    plt.title("std * n=100")
    plt.hist(dist, bins=100)
    plt.show()

    plt.xlabel = "n"
    plt.ylabel = "std"
    plt.yticks(np.array(range(0, 11)) / 10)
    plt.ylim(0, 1)
    plt.plot(list(range(1, m)), stds);
    
plot_std_reduction(lambda n: np.random.normal(size=(10000, n)))
```

# Beta

```python
dist = np.random.beta(0.5, 0.5, size=10000)
print("std", np.std(dist))
plt.hist(dist, bins=100);
```

```python
plot_std_reduction(lambda n: np.random.beta(0.5, 0.5, size=(10000, n)))
```

# Exponential

```python
dist = np.random.pareto(2, size=10000)
print("std", np.std(dist))
plt.hist(dist, bins=100);
```

```python
plot_std_reduction(lambda n: np.random.pareto(2, size=(10000, n)))
```
