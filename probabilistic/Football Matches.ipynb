# Building dataset

```python
with open("uefa-fixtures.pkl",'rb') as f:
    fixtures = pickle.load(f)["api"]["fixtures"]

with open("lineups.pkl",'rb') as f:
    lineups = pickle.load(f)

with open("events.pkl",'rb') as f:
    events = pickle.load(f)
    
with open("uefa-fixtures-2018.pkl",'rb') as f:
    fixtures += pickle.load(f)["api"]["fixtures"]
    
with open("lineups-2018.pkl",'rb') as f:
    lineups.update(pickle.load(f))

with open("events-2018.pkl",'rb') as f:
    events.update(pickle.load(f))
```

```python
import pandas as pd

data = {
    "match_id": [],
    "match_date": [],
    "player": [],
    "player_id": [],
    "team": [],
    "team_id": [],
    "position": [],
    "goals": [],
}
for fixture in fixtures:
    fixture_id = fixture['fixture_id']
    lineup = lineups[fixture_id]
    eventlist = events[fixture_id] 
    goals = {}
    substituitions = []
    for event in eventlist:
        if event['type'] == 'Goal':
            player_id = event['player_id']
            if player_id in goals:
                goals[player_id] += 1
            else:
                goals[player_id] = 1
        elif event['type'] == 'subst':
            substituitions.append(event['assist_id'])
            
    for team in lineup.keys():
        players = lineup[team]['startXI']
        players += [ x for x in lineup[team]['substitutes'] if x['player_id'] in substituitions ]
        for player in players:
            player_id = player["player_id"]
            goals_count = goals[player_id] if player_id in goals else 0
            
            data["match_id"].append(fixture_id)
            data["match_date"].append(fixture["event_date"])
            data["player"].append(player["player"])
            data["player_id"].append(player_id)
            data["team"].append(team)
            data["team_id"].append(player["team_id"])
            data["position"].append(player["pos"])
            data["goals"].append(goals_count)

df = pd.DataFrame(data)
df.head()
```
